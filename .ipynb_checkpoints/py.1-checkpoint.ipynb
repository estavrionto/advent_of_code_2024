```python
t = '''3   4
4   3
2   5
1   3
3   9
3   3'''
def d1(df_t):
    filter = ['r1','r2']
    df_t.columns = filter
    df_t['s1'] = df_t['r1'].sort_values().values

    # df_t['s1'] = df_t.sort_values(by=['r1'])['r1'].values
    # display(df_t.sort_values(by=['r1']))
    # display(df_t.sort_values(by=['r1'])['r1'])
    df_t['s2'] = df_t['r2'].sort_values().values
    df_t['d1'] = df_t['s2']-df_t['s1']
    df_t['d2'] = df_t['d1'].abs()
    display(df_t)
    print(df_t['d2'].sum())
    return df_t

def d2(df_t):
    l_t = []
    for index,row in df_t.iterrows():
        i = row['r1']
        # print(i)
        filter = df_t['r2']==i
        df_tt = df_t[filter]
        # display(df_tt)
        j = df_tt.shape[0]
        # print(i,j)
        l_t.append(i*j)
    df_t['D2.i*j'] = l_t
    print(df_t['D2.i*j'].sum())
    display(df_t)

    return df_t

    
t = t.split('\n')
df_t = [[],[]]
for i in t:
    i = i.split(' ')
    print(i)
    df_t[0].append(i[0])
    df_t[-1].append(i[-1])
df_t = pd.DataFrame(df_t).T
df_t = df_t.astype(int)
df_t = d1(df_t)
df_t = d2(df_t)
```

```python
def check_num(n):
    if len(n)>3:
        return 0
    digits = '0123456789'
    for i in n:
        if i not in digits:
            return 0
    return int(n)
        
    return n
def day3_1(s):
    # print(s)
    s = s.split('mul(')
    summ = 0
    return_l = []
    for i in s:
        # print(i)
        if ',' not in i:
            continue
        ii = i.split(',')
        if len(ii)<2:
            continue
        n1 = ii[0]
        n1 = check_num(n1)
        n2 = ii[1]
        if ')' not in n2:
            continue
        n2 = n2.split(')')[0]
        n2 = check_num(n2)
        # print(f'mul({n1},{n2})')
        if 0 not in [n1,n2]:
            return_l.append([n1,n2])
            summ += n1*n2

        # print([i,n1,n2])
    return summ, return_l

# df_3 = 'xmul(2,4)%&mul[3,7]!@^do_not_mul(5,5)+mul(32,64]then(mul(11,8)mul(8,5))'

# df_3 = "xmul(2,4)&mul[3,7]!^don't()_mul(5,5)+mul(32,64](mul(11,8)undo()?mul(8,5))"
# df_3 = df_3.split('\n')
# df_3 = ['do()'+i for i in df_3]
# df_3 = ''.join(df_3)
# print(df_3)


with open('./input.day_003.txt') as f:
    df_3 = f.read()
len(df_3)

result1, l_t1 = day3_1(df_3)
print(result1)

# df_3 = df_3.split('\n')
# df_3 = ['do()'+i for i in df_3]
# df_3 = ''.join(df_3)

df_3 = 'do()'+df_3

result3_p2 = 0
for i in df_3.split("don't()"):
    if 'do()' not in i:
        continue
    ii = i.split('do()')[1:]
    # print(ii)
    ii = ''.join(ii)
    # print(ii)
    r3_p2, l_t3_p2 = day3_1(ii)
    result3_p2 += r3_p2
print(result3_p2)
```
