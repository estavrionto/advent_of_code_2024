```python
import pandas as pd
df_t = pd.read_csv('./input.txt',sep='   ',engine='python',header=None)
df_t = d1(df_t)
df_t = d2(df_t)
```

```python
t = '''3   4
4   3
2   5
1   3
3   9
3   3'''
def d1(df_t):
    filter = ['r1','r2']
    df_t.columns = filter
    df_t['s1'] = df_t['r1'].sort_values().values

    # df_t['s1'] = df_t.sort_values(by=['r1'])['r1'].values
    # display(df_t.sort_values(by=['r1']))
    # display(df_t.sort_values(by=['r1'])['r1'])
    df_t['s2'] = df_t['r2'].sort_values().values
    df_t['d1'] = df_t['s2']-df_t['s1']
    df_t['d2'] = df_t['d1'].abs()
    display(df_t)
    print(df_t['d2'].sum())
    return df_t
t = t.split('\n')
df_t = [[],[]]
for i in t:
    i = i.split(' ')
    print(i)
    df_t[0].append(i[0])
    df_t[-1].append(i[-1])
df_t = pd.DataFrame(df_t).T
df_t = df_t.astype(int)
df_t = d1(df_t)
df_t = d2(df_t)
```

```python
def d2(df_t):
    l_t = []
    for index,row in df_t.iterrows():
        i = row['r1']
        # print(i)
        filter = df_t['r2']==i
        df_tt = df_t[filter]
        # display(df_tt)
        j = df_tt.shape[0]
        # print(i,j)
        l_t.append(i*j)
    df_t['D2.i*j'] = l_t
    print(df_t['D2.i*j'].sum())
    display(df_t)

    return df_t
d2(df_t)
```
