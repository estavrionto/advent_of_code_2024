```python
# d17
d17_exp='day_017'
with open(f'./input.{d17_exp}.txt') as f:
    l_d17 = f.read()
l_d17 = str(l_d17)


class d17_C:
    def __init__(self, str_i,log=False):
        self.p_print(sep,log=log)
        str_i = str_i.split('\n')
        str_i = [i for i in str_i if i!= '']
        str_i = [i.split(': ') for i in str_i]
        str_i = {i[0]:[int(j) for j in i[1].split(',')] for i in str_i}
        # print(str_i)
        # program = []
        # row=[]
        # for n,i in enumerate(str_i['Program']):
        #     if n%2==0:
        #         row.append(i)
        #     elif n%2==1:
        #         row.append(i)
        #         program.append(row)
        #         row=[]
        # print(program)
                
        
        # for i in 'ABC':
        #     l_d17[i] = l_d17[i][0]
      
        self.rA = str_i['Register A'][0]
        self.rB = str_i['Register B'][0]
        self.rC = str_i['Register C'][0]
        self.Program = str_i['Program']
        self.instruction_pointer=0
        self.out=[]
        self.log=log
        self.break_main_loop = False
        self.df_all = []
        self.row = []
        self.opcode_name = 'INIT'
        t = [str(i) for i in self.Program]
        self.str_program = ','.join(t)
        # print(self)

        
        

    def __str__(self):
        p_str = ''
        p_str = p_str+f'A: {self.rA}\n'
        p_str = p_str+f'B: {self.rB}\n'
        p_str = p_str+f'C: {self.rC}\n'
        p_str = p_str+f'Program: {self.Program}\n'
        
        return p_str
    def interpret_CO(self,CO_literal):
        if   CO_literal==0:CO=0
        elif CO_literal==1:CO=1
        elif CO_literal==2:CO=2
        elif CO_literal==3:CO=3
        elif CO_literal==4:CO=self.rA
        elif CO_literal==5:CO=self.rB
        elif CO_literal==6:CO=self.rC
        elif CO_literal==7:raise ValueError(f"literal CO {CO_literal}")
        return CO
    def opcode(self,OP,CO_literal):
        if OP==0:self.opcode_0(OP,CO_literal)
        elif OP==1:self.opcode_1(OP,CO_literal)
        elif OP==2:self.opcode_2(OP,CO_literal)
        elif OP==3:self.opcode_3(OP,CO_literal)
        elif OP==4:self.opcode_4(OP,CO_literal)
        elif OP==5:self.opcode_5(OP,CO_literal)
        elif OP==6:self.opcode_6(OP,CO_literal)
        elif OP==7:self.opcode_7(OP,CO_literal)
    def name_opcode(self,OP):
        dict_opcode = {
            0:'DIV_of_A/CO:___A',
            1:'XOR_of_B/CO_l:_B',
            2:'MOD_of_CO:_____B',
            3:'JUMP_to_CO:____0',
            4:'XOR_of_B/C:____B',
            5:'OUTPUT:______out',
            6:'DIV_of_A/CO:___B',
            7:'DIV_of_A/CO:___C',
        }
        return dict_opcode[OP]
            
    def opcode_0(self,OP,CO_literal):
        self.opcode_name = 'adv'
        CO = self.interpret_CO(CO_literal)
        self.p_print(f'[opcode_{OP}:{self.opcode_name} start] CO literal: {CO_literal}')
        self.p_print(f'[opcode_{OP}:{self.opcode_name} start] CO: {CO}')
        rt = self.rA/2**CO
        rt = int(rt)
        self.rA = rt
        self.update_ip(self.instruction_pointer+2)
        self.p_print(f'[opcode_{OP}:{self.opcode_name} end]')

        # self.instruction_pointer+=2
    def opcode_1(self,OP,CO_literal):
        self.opcode_name = 'bxl'
        # CO = self.interpret_CO(CO_literal)
        self.p_print(f'[opcode_{OP}:{self.opcode_name} start] CO literal: {CO_literal}')
        # self.p_print(f'[opcode_{OP}:{self.opcode_name} start] CO: {CO}')
        rt = self.rB^CO_literal
        self.rB = rt
        self.update_ip(self.instruction_pointer+2)
        # self.instruction_pointer+=2
        self.p_print(f'[opcode_{OP}:{self.opcode_name} end]')

    def opcode_2(self,OP,CO_literal):
        self.opcode_name = 'bst'
        CO = self.interpret_CO(CO_literal)
        self.p_print(f'[opcode_{OP}:{self.opcode_name} start] CO literal: {CO_literal}')
        self.p_print(f'[opcode_{OP}:{self.opcode_name} start] CO: {CO}')
        rt = CO%8
        self.rB = rt
        self.update_ip(self.instruction_pointer+2)
        # self.instruction_pointer+=2
        self.p_print(f'[opcode_{OP}:{self.opcode_name} end]')

    def opcode_3(self,OP,CO_literal):
        self.opcode_name = 'jnz'
        CO = self.interpret_CO(CO_literal)
        self.p_print(f'[opcode_{OP}:{self.opcode_name} start] CO literal: {CO_literal}')
        self.p_print(f'[opcode_{OP}:{self.opcode_name} start] CO: {CO}')
        if self.rA != 0:
            self.update_ip(CO)
            # self.instruction_pointer=CO
        else:
            self.update_ip(self.instruction_pointer+2)
            # self.instruction_pointer+=2
        self.p_print(f'[opcode_{OP}:{self.opcode_name} end]')

    
    def opcode_4(self,OP,CO_literal):
        self.opcode_name = 'bxc'
        CO = self.interpret_CO(CO_literal)
        self.p_print(f'[opcode_{OP}:{self.opcode_name} start] CO literal: {CO_literal}')
        self.p_print(f'[opcode_{OP}:{self.opcode_name} start] CO: {CO}')
        rt = self.rB^self.rC
        self.rB = rt
        self.update_ip(self.instruction_pointer+2)
        # self.instruction_pointer+=2
        self.p_print(f'[opcode_{OP}:{self.opcode_name} end]')

    
    def opcode_5(self,OP,CO_literal):
        self.opcode_name = 'out'
        CO = self.interpret_CO(CO_literal)
        self.p_print(f'[opcode_{OP}:{self.opcode_name} start] CO literal: {CO_literal}')
        self.p_print(f'[opcode_{OP}:{self.opcode_name} start] CO: {CO}')
        rt = CO%8
        self.out.append(rt)
        self.update_ip(self.instruction_pointer+2)
        # self.instruction_pointer+=2
        self.p_print(f'[opcode_{OP}:{self.opcode_name} end]')

    
    def opcode_6(self,OP,CO_literal):
        self.opcode_name = 'bdv'
        CO = self.interpret_CO(CO_literal)
        self.p_print(f'[opcode_{OP}:{self.opcode_name} start] CO literal: {CO_literal}')
        self.p_print(f'[opcode_{OP}:{self.opcode_name} start] CO: {CO}')
        rt = self.rA/2**CO
        rt = int(rt)
        self.rB = rt
        self.update_ip(self.instruction_pointer+2)
        self.p_print(f'[opcode_{OP}:{self.opcode_name} end]')

    
    def opcode_7(self,OP,CO_literal):
        self.opcode_name = 'cdv'
        CO = self.interpret_CO(CO_literal)
        self.p_print(f'[opcode_{OP}:{self.opcode_name} start] CO literal: {CO_literal}')
        self.p_print(f'[opcode_{OP}:{self.opcode_name} start] CO: {CO}')
        rt = self.rA/2**CO
        rt = int(rt)
        self.rC = rt
        self.update_ip(self.instruction_pointer+2)
        self.p_print(f'[opcode_{OP}:{self.opcode_name} end]')

    def update_ip(self,new_value):
        self.instruction_pointer = new_value
        self.p_print(f'[update ip] Program: {self.Program}, len: {len(self.Program)}')
        if new_value >= len(self.Program):
            self.break_main_loop = True
        
    def p_status0(self,p_str='',log=None):
        if log==None:
            log = self.log
        if log:
            p_str = p_str+f'A: {self.rA}, '
            p_str = p_str+f'B: {self.rB}, '
            p_str = p_str+f'C: {self.rC}, ' 
            p_str = p_str+f'Program: {self.Program}, '
            p_str = p_str+f'out: {self.out}, '
            p_str = p_str+f'instruction_pointer: {self.instruction_pointer}, '
            p_str = p_str+f'break_main_loop: {self.break_main_loop} '
            p_str = p_str+f'log: {self.log} '
            self.p_print(p_str,log=log)

    
    def p_status(self,p_str='',log=None):
        if log==None:
            log = self.log
        if log:
            p_str = p_str+f'A: {self.rA}, '
            p_str = p_str+f'B: {self.rB}, '
            p_str = p_str+f'C: {self.rC}, ' 
            p_str = p_str+f'out: {self.out}, '
            p_str = p_str+f'instruction_pointer: {self.instruction_pointer}, '
            # p_str = p_str+f'Program: {self.Program}\n'
            self.p_print(p_str,log=log)
    def p_print(self,p_str='',log=None):
        if log==None:
            log = self.log
        if log:
            print(p_str)
    def p_output(self,ret_str=False):
        p_str = [str(i) for i in self.out]
        p_str = ','.join(p_str)
        if ret_str:
            return p_str
        else:
            self.p_print(p_str,log=True)
    def p_display(self,df='',log=None):
        if log==None:
            log = self.log
        if log:
            display(df)

        
    def main(self,log=None):
        if log!=None:
            self.log=log
        # print(self)
        self.p_print(sep,log=log)
        self.p_status0(f'[main, start] ',log=log)
        i=-1
        while True:
            i+=1
            if self.break_main_loop == True:
                break
            
            self.p_print(sep,log=log)
            self.p_status(f'[main loop, iter {i} start] ')
            in_OP=self.Program[self.instruction_pointer]
            in_CO=self.Program[self.instruction_pointer+1]
            self.row = [self.rA,self.rB,self.rC,in_OP,self.name_opcode(in_OP),in_CO,self.interpret_CO(in_CO),self.p_output(ret_str=True),self.instruction_pointer]
            self.df_all.append(self.row)

            self.p_print(f'[main loop, iter {i}] opcode: {in_OP}, literal CO:{in_CO}')
            self.opcode(in_OP,in_CO)
            self.p_status(f'[main loop, iter {i} end] ',log=log)
            
        self.row = [self.rA,self.rB,self.rC,in_OP,self.name_opcode(in_OP),in_CO,self.interpret_CO(in_CO),self.p_output(ret_str=True),self.instruction_pointer]
        self.df_all.append(self.row)
        self.df_all = pd.DataFrame(self.df_all,columns=[c for c in 'rA,rB,rC,OP,fun,CO_l,CO,out,iP'.split(',')])
        # t = [str(i) for i in self.Program]
        # self.p_print(','.join(t),log=True)
        # self.p_display(self.df_all,log=True)
        self.p_print(sep,log=log)
        self.p_status0(f'[main, end  ] ',log=log)
        self.p_print(sep,log=log)
        return self.str_program, self.df_all,self.p_output(ret_str=True)
        # self.p_output()



# main
ll_d17 = [
    # "|Register A: 729|Register B: 0|Register C: 0|Program: 0,1,5,4,3,0",
    # "|Register A: 0|Register B: 0|Register C: 9|Program: 2,6",
    # "|Register A: 10|Register B: 0|Register C: 0|Program: 5,0,5,1,5,4",
    # "|Register A: 2024|Register B: 0|Register C: 0|Program: 0,1,5,4,3,0",
    # "|Register A: 0|Register B: 29|Register C: 0|Program: 1,7",
    # "|Register A: 0|Register B: 2024|Register C: 43690|Program: 4,0",
    # "|Register A: 38610541|Register B: 0|Register C: 0|Program: 2,4,1,1,7,5,1,5,4,3,5,5,0,3,3,0",


    
    # "|Register A: 2024|Register B: 0|Register C: 0|Program: 0,3,5,4,3,0",
    "|Register A: 117440|Register B: 0|Register C: 0|Program: 0,3,5,4,3,0",


]
for l in ll_d17:
    l = l.replace('|','\n')
    # print(f'"{l}",')
    oi_d17 = d17_C(l)
    
    # oi_d17.main(log=True)
    oi_d17.main(log=False)
    # break
```
