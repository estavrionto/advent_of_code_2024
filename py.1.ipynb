```python
%%time
# with open('./input.day_006.e.txt') as f:
# with open('./input.day_006.e1.txt') as f:
with open('./input.day_006.txt') as f:
    l_d6 = f.read()
def d6_find(l,str_in):
    # d6_display(l)
    # direction = ''
    pos = []
    for r,row in enumerate(l):
        for c,col in enumerate(row):
            if col in str_in:
                pos.append([r,c])
                
            # if col=='^':
            #     direction='^'
            # elif col=='v':
            #     direction='v'
            # elif col=='>':
            #     direction='>'
            # elif col=='<':
            #     direction='<'
            # if direction!='':
                # print(r,c,direction)
                # return r,c
    return pos
l_d6 = str(l_d6)
l_d6 = l_d6.split('\n')
l_d6 = [i for i in l_d6 if i!= '']
l_d6 = [[j for j in i] for i in l_d6]

d6_remap = {
    '^':'>',
    '>':'v',
    'v':'<',
    '<':'^',
}
d6_map = {
    '^':'↑',
    '>':'→',
    'v':'↓',
    '<':'←',
}

    # '↑→↓←'
def d6_display(l_of_l,plot=True):
    if plot==True:
        l_str = []
        for i in range(len(l_of_l[0])):
            for l in l_of_l:
                r = ''.join(l[i])
                print(r,end='\t')
            print()
        print(sep)
def d6_update(Map,row,col,Map_c,turns,plot='end'):
    if type(Map_c) == str:
        Map_c = [['0' for c in r] for r in Map]
    if type(turns) == str:
        turns = []
    if plot=='end':
        p_1 = True
        p_2 = False
        p_3 = False
    elif plot=='turns':
        p_1 = True
        p_2 = True
        p_3 = False
    elif plot=='all':
        p_1 = True
        p_2 = True
        p_3 = True
    else:
        p_1 = False
        p_2 = False
        p_3 = False

    try:
        drc = Map[row][col]
    except:
        print('problem')
        d6_display([Map])
        raise ValueError(f"{row},{col}")
    row_new = copy.deepcopy(row)
    col_new = copy.deepcopy(col)
    # Map[row][col] = 'X'
    if drc == '^':
        row_new = row-1
    elif drc == 'v':
        row_new = row+1
    elif drc == '<':
        col_new = col-1
    elif drc == '>':
        col_new = col+1
    else:
        
        raise ValueError(f"{drc} not a ship")


    status = 'continue'
    global known_path
    fill='X'
    fill=d6_map[drc]
    if row_new not in range(len(Map)):
        # end of map
        Map[row][col] = fill
        Map_c = d6_update_path(Map_c,row,col)
        status='end'
        d6_display([Map,Map_c],plot=p_1)
    elif col_new not in range(len(Map[0])):
        # end of map
        Map[row][col] = fill
        Map_c = d6_update_path(Map_c,row,col)
        status='end'
        d6_display([Map,Map_c],plot=p_1)
    elif Map[row_new][col_new] in '#O':
        # reach wall
        
        drc = d6_remap[drc]
        Map[row][col] = drc
        row_new = copy.deepcopy(row)
        col_new = copy.deepcopy(col)
        d6_display([Map,Map_c],plot=p_2)
        status = 'turn'
        turns.append([row,col,drc])
        # if d6_check_inf(copy.deepcopy(Map),row,col,drc,turns) =='loop':
        #     # status='turn_loop_end'
        #     status='turn_loop'
    else:
        Map[row][col] = fill
        Map_c = d6_update_path(Map_c,row,col)
        Map[row_new][col_new] = drc
        d6_display([Map,Map_c],plot=p_3)
        status = 'forward'
    return Map, row_new, col_new, status, Map_c, turns
    
def d6_update_path(Map_c,r,c):
    cur = Map_c[r][c]
    cur = int(cur)+1
    Map_c[r][c] = str(cur)
    return Map_c

def d6_newwall(turns3,drc):
    if drc=='<':
        r = turns3[2][0]
        c = turns3[0][1]-1
    if drc=='^':
        r = turns3[0][0]-1
        c = turns3[2][1]
    if drc=='>':
        r = turns3[2][0]
        c = turns3[0][1]+1
    if drc=='v':
        r = turns3[0][0]+1
        c = turns3[2][1]
    return r,c

def d6_map_replace(Map,str_in,char_out):
    for i,r in enumerate(Map):
        for j,c in enumerate(r):
            x = Map[i][j]
            if x in str_in:
                x=char_out
            Map[i][j] = x
    return Map
def d6_check_inf(Map_t,row,col,drc,wall_row,wall_col):
    # nw_r,nw_c = d6_newwall(tr,drc)
    Map = copy.deepcopy(Map_t)
    # print(row,col,drc,wall_row,wall_col)
    Map[wall_row][wall_col]='O'
    Map[row][col] = drc
    
    # print([row,col,drc,turns])
    # print(Map[row][col])
    # Map[0][0] = '#'
    Map = d6_map_replace(Map,'X↑→↓←','.')

    # print(type(row),row)
    # Map[row][col] = 'a'
    # print(row,col,drc)
    # print('fun_inf: new map')
    # d6_display([Map])
    mc = ''
    t = ''
    MapB=''
    loop_c=0
    for i in range(100000):
        # print(sep,i)
        Map,row,col,status,mc,t = d6_update(Map,row,col,'','',plot='None')
        
        if 'turn' in status:
            if d6_map_replace(copy.deepcopy(Map),'X↑→↓←^>v<','X') == MapB:
                loop_c+=1
                # print('loop_c increase')
            MapB = d6_map_replace(copy.deepcopy(Map),'X↑→↓←^>v<','X')
            # d6_display([Map,MapB])
        if 'end' in status:
            break
        if loop_c>50:
            # print('fun_inf: inf found')
            # d6_display([Map,MapB])
            # print(sep)
            return 'loop'
    # print('fun_inf: inf not found',)
    # d6_display([Map,])
    # print(sep)
    return 'no_loop'            
    # else:
    #     # print('fun_inf: not enough turns')
    #     # print(sep)
    #     return 'no_loop'

def d6_find_new_walls(Map,t0,t1=None,not_valid=[],log=False):
    if log:
        p = print
    else:
        def p(*args):
            pass
    max_r = len(Map)-1
    max_c = len(Map[0])-1
    p('max R,C:',max_r,max_c)
    p('points:',t0,t1)
    if t0[2]=='>':
        if t1==None:
            t1 = [max_r,max_c]
        R = [t0[0]]
        C = [i+1 for i in range(t0[1],t1[1])]
        search_R = [i+1 for i in range(t0[0],max_r)]
        search_C = [i+1 for i in range(t0[1],t1[1])]
    if t0[2]=='<':
        if t1==None:
            t1 = [0,0]
        R = [t0[0]]
        C = [i for i in range(t1[1],t0[1])]
        search_R = [i for i in range(0,t0[0])]
        search_C = [i for i in range(t1[1],t0[1])]
    if t0[2]=='v':
        if t1==None:
            t1 = [max_r,0]
        R = [i for i in range(t0[0],t1[0]+1)]
        C = [t0[1]]
        search_R = [i for i in range(t0[0],t1[0]+1)]
        search_C = [i for i in range(0,t0[1])]
    if t0[2]=='^':
        if t1==None:
            t1 = [0,max_c]
        R = [i for i in range(t1[0],t0[0])]
        C = [t0[1]]
        search_R = [i for i in range(t1[0],t0[0])]
        search_C = [i+1 for i in range(t0[1],max_c)]

    p('turn field R,C:',R,C)
    p('search field R,C',search_R,search_C)
    search_RC = []

    all_walls = d6_find(Map,'#O^>v<')
    p('all_walls:',all_walls)

    for sr in search_R:
        for sc in search_C:
            search_RC.append([sr,sc])
    p('search field RC:',search_RC)
    search_RC = [i for i in search_RC if i in all_walls]
    p('walls found:',search_RC)
                
    RC = []
    for r in R:
        for c in C:
            RC.append([r,c])
    p('possible turns:',RC)
    new_turns_RC = []
    for rc in RC:
        for s_rc in search_RC:
            if rc[0]==s_rc[0]:
                new_turns_RC.append(rc)
            if rc[1]==s_rc[1]:
                new_turns_RC.append(rc)
    p('new turns',new_turns_RC)
    p('not valid:',not_valid)
    if t0[2]=='>':
        new_walls_RC = [[i[0],i[1]+1] for i in new_turns_RC]
    if t0[2]=='<':
        new_walls_RC = [[i[0],i[1]-1] for i in new_turns_RC]
    if t0[2]=='v':
        new_walls_RC = [[i[0]+1,i[1]] for i in new_turns_RC]
    if t0[2]=='^':
        new_walls_RC = [[i[0]-1,i[1]] for i in new_turns_RC]
    p('new_walls_RC:',new_walls_RC)

    new_walls_RC = [i for i in new_walls_RC if i not in not_valid]
    new_walls_RC = [i for i in new_walls_RC if i[0] <= max_r]
    new_walls_RC = [i for i in new_walls_RC if i[1] <= max_c]
    p('new_walls_RC valid:',new_walls_RC)
    p(sep)
    return new_walls_RC
                
                

def d6_main(Map):
    row,col = d6_find(Map,'^>v<')[0]
    print(row,col)
    not_valid = d6_find(Map,'#O^>v<')
    Map_c = ''
    turns = ''
    statuses = []
    for i in range(10000):
        # print(sep,i)
        Map,row,col,status,Map_c,turns = d6_update(Map,row,col,Map_c,turns,
                                                   # plot='turns',
                                                   plot='None',
                                                  )
        statuses.append(status)
        if 'end' in status:
            break
    if status=='end':
        count = 0
        for r in Map:
            for c in r:
                if c in 'X↑→↓←':
                    count+=1
        print('final count:',count)
        print('num turns',len(turns))
        # count = 0
        # for i in statuses:
        #     if i == 'turn_loop':
        #         count+=1
        # print('d6 p2:',count)
        d6_display([Map,Map_c])
        all_walls = []
        for i,t in enumerate(turns):
            if i+1 == len(turns):
                t1=None
            else:
                t1=turns[i+1]
            # if t[2]!='v':
            #     continue
            walls = d6_find_new_walls(Map,t,t1,not_valid)
            # print(t,walls)
            [all_walls.append([t,i]) for i in walls]
        count=0
        print(sep)
        print('num walls valid to add:',len(all_walls))
        time = 0
        for start,wall in all_walls:
        # for start,wall in [[[6, 8, '<'], [6, 3]]]:
            # print(start,wall)
            time +=1
            print(time,time%100)
            if time%100==0:
                print(time)
            st = d6_check_inf(Map,start[0],start[1],start[2],wall[0],wall[1])
            if st=='loop':
                count+=1
                # break
            # break
        print('final loop count:',count)


    else:
        print('increase loop')
d6_main(l_d6)
```
